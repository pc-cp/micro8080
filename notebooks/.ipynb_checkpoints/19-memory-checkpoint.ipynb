```python
class Decade_Counter_10:
    """A module-10 Ripple Counter that counts from 0 to 9 and resets. """
    def __init__(self,):
        self.nbits = 4
        self.flipflops = [Edge_Triggered_D_Flip_Flop_V2() for _ in range(self.nbits)]
        self.and_gate = AND()

    def getQs(self):
        qs = [ff.getQ() for ff in self.flipflops]
        return list(reversed(qs))
    
    def __call__(self, clk = 0):
        # The hardware stabilization loop
        while True:
            # 1. Take a snapshot of the current state of the ff
            old_qs = [ff.getQ() for ff in self.flipflops]

            # 2. Wire up the Async Clear
            # ff[0] is Q0 (LSB), ff[1] is Q1, ff[2] is Q2, ff[3] is Q3 (MSB)
            # We want to clear when we hit 10 (Binary: 1010). 
            # This happens when Q3 == 1 and Q1 == 1.
            q1 = self.flipflops[1].getQ()
            q3 = self.flipflops[3].getQ()
            clear_wire = self.and_gate([q3, q1])

            # 3. Propagate the clock and data through the ripple chain
            current_clk = clk
            for ff in self.flipflops:
                data = ff.getQ_bar()
                # Pass Data, Clock, Preset(0), and our dynamic Clear wire!
                q, q_bar = ff([data, current_clk, 0, clear_wire])
                current_clk = q_bar

            # 4. Read the new state
            new_qs = [ff.getQ() for ff in self.flipflops]

            # 5. If the electrons have settled and nothing changed, break!
            if old_qs == new_qs:
                break

        # Return with MSB on the left for easy reading
        return list(reversed(new_qs))
        
        # current_clk = clk
        # qs = []
        # for flipflop in self.flipflops:
        #     data = flipflop.getQ_bar()
        #     q, q_bar = flipflop([data, current_clk])
        #     qs.append(q)
        #     current_clk = q_bar
        # # MSB is on the left
        # qs = list(reversed(qs))
        # # when the output of the four
        # # flip-flop is 1010 - which is 10
        # # we want all flip-flop cleared to zero.
        # clear = self.and_gate([qs[0], qs[2]])

        # current_clk = clk
        # qs = []
        # for flipflop in self.flipflops:
        #     data = flipflop.getQ_bar()
        #     q, q_bar = flipflop([data, current_clk, 0, clear])
        #     qs.append(q)
        #     current_clk = q_bar
        
        # return list(reversed(qs))

```

```python
class Decade_Counter_6:
    """A module-6 Ripple Counter that counts from 0 to 5 and resets. """
    def __init__(self,):
        self.nbits = 3
        self.flipflops = [Edge_Triggered_D_Flip_Flop_V2() for _ in range(self.nbits)]
        self.and_gate = AND()
        
    def getQs(self):
        qs = [ff.getQ() for ff in self.flipflops]
        return list(reversed(qs))
        
    def __call__(self, clk = 0):
        # when the while loop runs a second time, it means
        # the circuit hasn't settled yet, but we are still
        # inside the exact same moment in time.
        # By putting current_clk = clk inside the top of 
        # the while loop, you ensure that every time you
        # re-calculate the ripple chain, you are correctly
        # starting from the external physical pin,
        # rather than accidentally passing the baton in a circle.
        # current_clk = clk

        
        # The hardware stabilization loop
        while True:
            # 1. Take a snapshot of the current state of the ff
            old_qs = [ff.getQ() for ff in self.flipflops]

            # 2. Wire up the Async Clear
            # ff[0] is Q0 (LSB), ff[1] is Q1, ff[2] is Q2(MSB)
            # We want to clear when we hit 6 (Binary: 110). 
            # This happens when Q2 == 1 and Q1 == 1.
            q1 = self.flipflops[1].getQ()
            q2 = self.flipflops[2].getQ()
            clear_wire = self.and_gate([q2, q1])

            # 3. Propagate the clock and data through the ripple chain
            current_clk = clk
            for ff in self.flipflops:
                data = ff.getQ_bar()
                # Pass Data, Clock, Preset(0), and our dynamic Clear wire!
                q, q_bar = ff([data, current_clk, 0, clear_wire])
                current_clk = q_bar

            # 4. Read the new state
            new_qs = [ff.getQ() for ff in self.flipflops]

            # 5. If the electrons have settled and nothing changed, break!
            if old_qs == new_qs:
                break
            # else:
            #     print(list(reversed(old_qs)))
            #     print(list(reversed(new_qs)))
            # 5 -> 0
            # [1, 0, 1] 5
            # [1, 1, 0] 6
            # [1, 1, 0] 6
            # [0, 0, 0] 0

        # Return with MSB on the left for easy reading
        return list(reversed(new_qs))
```

```python
class Second_Timer:
    def __init__(self):
        self.decade_counter_6 = Decade_Counter_6()
        self.decade_counter_10 = Decade_Counter_10()
        self.nand_gate = NAND()

    def getBitsOfLow(self):
        bits_of_low = self.decade_counter_10.getQs()
        return bits_of_low

    def getBitsOfHigh(self):
        bits_of_high = self.decade_counter_6.getQs()
        return bits_of_high
    
    def __call__(self, clk = 0):
        bits_of_digit10 = self.decade_counter_10(clk)
        clk_of_digit6 = self.nand_gate([bits_of_digit10[0], bits_of_digit10[-1]])
        bits_of_digit6 = self.decade_counter_6(clk_of_digit6)
        return bits_of_digit6, bits_of_digit10
```

```python
class Minute_Timer:
    def __init__(self):
        self.decade_counter_6 = Decade_Counter_6()
        self.decade_counter_10 = Decade_Counter_10()
        self.nand_gate = NAND()

    def getBitsOfLow(self):
        bits_of_low = self.decade_counter_10.getQs()
        return bits_of_low

    def getBitsOfHigh(self):
        bits_of_high = self.decade_counter_6.getQs()
        return bits_of_high
    
    def __call__(self, clk = 0):
        bits_of_digit10 = self.decade_counter_10(clk)
        clk_of_digit6 = self.nand_gate([bits_of_digit10[0], bits_of_digit10[-1]])
        bits_of_digit6 = self.decade_counter_6(clk_of_digit6)
        return bits_of_digit6, bits_of_digit10
```

```python
class Minute_Second_Timer:
    def __init__(self):
        self.minute_timer = Minute_Timer()
        self.second_timer = Second_Timer()
        self.nand_gate = NAND()

    def __call__(self, clk = 0):
        bits_tens_sec, bits_ones_sec = self.second_timer(clk)
        clk_of_min = self.nand_gate([bits_tens_sec[0], bits_tens_sec[-1]])
        bits_tens_min, bits_ones_min = self.minute_timer(clk_of_min)
        
        return bits_tens_min, bits_ones_min, bits_tens_sec, bits_ones_sec
```

```python
class Hour_Timer:
    def __init__(self):
        self.nbits = 5        
        self.flipflops = [Edge_Triggered_D_Flip_Flop_V2() for _ in range(self.nbits)]
                
        self.nand_gate = NAND()
        self.and_gate_10 = AND(2) # Detects the 10
        self.and_gate_12 = AND(2) # Detects the 12
        
        self.or_gate = OR(2)

    def getQs(self):
        qs = [ff.getQ() for ff in self.flipflops]
        return list(reversed(qs))
    
    def __call__(self, clk=0):
        # The hardware stabilization loop
        while True:
            # 1. Take a snapshot of the current state of the ff
            old_qs = [ff.getQ() for ff in self.flipflops]
  
            # 2. Wire up the Async Clear
            q1 = self.flipflops[1].getQ()
            q3 = self.flipflops[3].getQ()
            q4 = self.flipflops[4].getQ()
            
            # Clear 1: When low digit hits 10 (q3=1, q1=1)
            clear_10 = self.and_gate_10([q3, q1])

            # Clear 2: When total time hits 12 (q4=1, q1=1). NO INVERTER NEEDED!
            clear_12 = self.and_gate_12([q4, q1])
            
            # The low digit must clear if it hits 10 OR if the whole clock hits 12
            clear_low_digit = self.or_gate([clear_12, clear_10])

            # The cascading clock for the High Digit (Triggers when low digit hits 10)
            q0 = self.flipflops[0].getQ()
            clk_of_hour_high_digit = self.nand_gate([q0, q3])
            
            # 3. Propagate the clock and data through the ripple chain
            current_clk = clk
            
            # Tick the low digit (bits 0 to 3)
            for ff in self.flipflops[:-1]:
                data = ff.getQ_bar()
                q, q_bar = ff([data, current_clk, 0, clear_low_digit])
                current_clk = q_bar

            # Tick the high digit (bit 4)
            data_high = self.flipflops[-1].getQ_bar()
            self.flipflops[-1]([data_high, clk_of_hour_high_digit, 0, clear_12])
            
            # 4. Read the new state
            new_qs = [ff.getQ() for ff in self.flipflops]

            # 5. If the electrons have settled and nothing changed, break!
            if old_qs == new_qs:
                break
                
        # Return with MSB on the left for easy reading
        return list(reversed(new_qs))
```

```python
# Power on the clock and our Hour Timer
system_clock = Oscillator()
hour_counter = Hour_Timer()

print("Tick | CLK |  High Low  | Decimal Hour")
print("-" * 38)

# Run for 30 ticks (15 simulated hours) to watch it roll over
for tick in range(1, 31):
    clk_signal = system_clock.level()
    
    # Run the hardware simulation
    bits = hour_counter(clk_signal)
    
    if clk_signal == 1:
        # Separate the High Digit (bit 0 of our reversed list) and Low Digit (bits 1-4)
        high_digit_bits = [bits[0]]
        low_digit_bits = bits[1:]
        
        val_high = bit_list_to_int(high_digit_bits, signed=False)
        val_low = bit_list_to_int(low_digit_bits, signed=False)
        
        # Format the hour
        print(f" {tick//2:2}  |  {clk_signal}  |   {val_high}   {val_low}   |     {val_high}{val_low}")
        
    system_clock.tick()
```

```python
class Hour_Minute_Second_Timer:
    def __init__(self):
        self.hour_timer = Hour_Timer()
        self.min_sec_timer = Minute_Second_Timer()
        
        self.nand_gate_hour_min = NAND()
        self.nand_gate_pm_am = NAND()
        # A toggle flip-flop to store AM (0) or PM (1)
        self.am_pm_memory = Edge_Triggered_D_Flip_Flop_V2()

    def __call__(self, clk = 0):
        bits_tens_min, bits_ones_min, bits_tens_sec, bits_ones_sec = self.min_sec_timer(clk)
        clk_of_hour = self.nand_gate_hour_min([bits_tens_min[0], bits_tens_min[-1]])
        bits_hour = self.hour_timer(clk_of_hour)
        bits_tens_hour, bits_ones_hour = [bits_hour[0]], bits_hour[1:] # The 1-bit high digit, The 4-bit low digit
        
        clk_of_am_pm = self.nand_gate_pm_am([bits_tens_hour[0], bits_ones_hour[-1]])
        current_am_pm_state = self.am_pm_memory.getQ_bar()
        am_pm_q, _ = self.am_pm_memory([current_am_pm_state, clk_of_am_pm, 0, 0])
        
        return am_pm_q, bits_tens_hour, bits_ones_hour, bits_tens_min, bits_ones_min, bits_tens_sec, bits_ones_sec
```

```python
# 1. Power on the System Clock and the full System Timer
system_clock = Oscillator()
master_clock = Hour_Minute_Second_Timer()

# 2. Open a text file in "write" mode
filename = "full_24hour_simulation.txt"
with open(filename, "w") as f:
    f.write("Tick | CLK | Time (HH:MM:SS AM/PM) | Raw Bits\n")
    f.write("-" * 80 + "\n")

    # 3. Simulate 86,405 seconds (A full 24-hour day, plus 5 seconds to watch it start over)
    total_ticks = 86405 * 2
    
    print(f"Starting simulation of {total_ticks} ticks. Please wait...")

    for tick in range(1, total_ticks + 1):
        clk_signal = system_clock.level()
        
        # Run the massive hardware simulation
        am_pm, t_hr, o_hr, t_min, o_min, t_sec, o_sec = master_clock(clk_signal)
        
        # Only record to file on the Rising Edge (CLK = 1)
        if clk_signal == 1:
            # Convert binary lists to human decimal
            val_t_hr = bit_list_to_int(t_hr, signed=False)
            val_o_hr = bit_list_to_int(o_hr, signed=False)
            val_t_min = bit_list_to_int(t_min, signed=False)
            val_o_min = bit_list_to_int(o_min, signed=False)
            val_t_sec = bit_list_to_int(t_sec, signed=False)
            val_o_sec = bit_list_to_int(o_sec, signed=False)
            
            # AM = 0, PM = 1
            period = "AM" if am_pm == 0 else "PM"
            
            # Note: A 0-11 clock traditionally displays "00" instead of "12" 
            time_str = f"{val_t_hr}{val_o_hr}:{val_t_min}{val_o_min}:{val_t_sec}{val_o_sec} {period}"
            
            # We don't want to print 100,000 lines, so we only write it to the file
            # unless it's the exact top of the hour!
            if val_t_min == 0 and val_o_min == 0 and val_t_sec == 0 and val_o_sec == 0:
                print(f"BONG! The time is exactly {time_str} (Tick {tick})")
            
            f.write(f" {tick:6} |  1  | {time_str:20} | {am_pm} {t_hr}{o_hr} {t_min}{o_min} {t_sec}{o_sec}\n")
            
        system_clock.tick()

print(f"\nSimulation complete! All 86,400 seconds saved to '{filename}'.")
```

# (Page 249:) when the clock start up, the initial time will be set to: 11:11:10m
The wire connecting the NAND gate to the next clock stage was at 0 Volts a millisecond ago, and is suddenly at 5 Volts. That sudden jump from 0V to 5V is a physical Rising Edge!
Because of this physical voltage spike:
- The Seconds Tens digit sees a rising edge and ticks to 1.
- The Minutes Ones digit sees a rising edge and ticks to 1.
- The Minutes Tens digit sees a rising edge and ticks to 1.
- The Hours Ones digit sees a rising edge and ticks to 1.
- The Hours Tens digit sees a rising edge and ticks to 1.
- The Seconds Ones digit is waiting on the external Oscillator, so it stays 0.

Result: 11:11:10.

Our Python implementation is mathematically perfect, but variables **do not have a "powered off" state**.
In our simulation, on the very first line of your loop (tick = 1), Python calculates NAND(0, 0) = 1 and instantly passes that 1 into the next flip-flop.
Because our Python flip-flop was handed a 1 on its very first evaluation, it never saw the transition. It just assumed the wire was always high. We essentially spawned the universe with the NAND gate already turned on, so the flip-flop never experienced the "Rising Edge."

How Engineers Fix This in Real Life
This 11:11:10 startup glitch is exactly why almost every digital chip in the world has a dedicated Power-On Reset (POR) circuit.

```python
class Hour_Minute_Second_Timer_V2:
    # adding switches to set the time manually(hour and minute and second)
    def __init__(self):
        self.hour_timer = Hour_Timer()
        self.minute_timer = Minute_Timer()
        self.second_timer = Second_Timer()
        self.nand_gate_min_sec = NAND()
        
        self.nand_gate_hour_min = NAND()
        self.nand_gate_pm_am = NAND()
        # A toggle flip-flop to store AM (0) or PM (1)
        self.am_pm_memory = Edge_Triggered_D_Flip_Flop_V2()

        self.switch_sec = XOR()
        self.switch_min = XOR()
        self.switch_hour = XOR()

    def update_hour(self, hour_stride):
        for _ in range(hour_stride):
            bits_tens_min = self.minute_timer.getBitsOfHigh()
            clk_of_hour = self.nand_gate_hour_min([bits_tens_min[0], bits_tens_min[-1]])
            for switch in [1, 1]:
                clk_of_hour = self.switch_hour([switch, clk_of_hour])
                bits_hour = self.hour_timer(clk_of_hour)
                bits_tens_hour, bits_ones_hour = [bits_hour[0]], bits_hour[1:] # The 1-bit high digit, The 4-bit low digit        
                clk_of_am_pm = self.nand_gate_pm_am([bits_tens_hour[0], bits_ones_hour[-1]])
                current_am_pm_state = self.am_pm_memory.getQ_bar()
                am_pm_q, _ = self.am_pm_memory([current_am_pm_state, clk_of_am_pm, 0, 0])

    def update_minute(self, minute_stride):
        for _ in range(hour_stride):
            bits_tens_sec = self.second_timer.getBitsOfHigh()
            clk_of_min = self.nand_gate_min_sec([bits_tens_sec[0], bits_tens_sec[-1]])
            for switch in [1, 1]:
                clk_of_min = self.switch_min([switch, clk_of_min])
                bits_tens_min, bits_ones_min = self.minute_timer(clk_of_min)
                
                clk_of_hour = self.nand_gate_hour_min([bits_tens_min[0], bits_tens_min[-1]])
                clk_of_hour = self.switch_hour([0, clk_of_hour])
                bits_hour = self.hour_timer(clk_of_hour)
                bits_tens_hour, bits_ones_hour = [bits_hour[0]], bits_hour[1:] # The 1-bit high digit, The 4-bit low digit
                
                clk_of_am_pm = self.nand_gate_pm_am([bits_tens_hour[0], bits_ones_hour[-1]])
                current_am_pm_state = self.am_pm_memory.getQ_bar()
                am_pm_q, _ = self.am_pm_memory([current_am_pm_state, clk_of_am_pm, 0, 0])

    def update_second(self, second_stride):
        for _ in range(second_stride):
            for switch in [1, 1]:
                clk = self.switch_sec([switch, clk])
                bits_tens_sec, bits_ones_sec = self.second_timer(clk)
                clk_of_min = self.nand_gate_min_sec([bits_tens_sec[0], bits_tens_sec[-1]])
                clk_of_min = self.switch_min([0, clk_of_min])
                bits_tens_min, bits_ones_min = self.minute_timer(clk_of_min)
                
                clk_of_hour = self.nand_gate_hour_min([bits_tens_min[0], bits_tens_min[-1]])
                clk_of_hour = self.switch_hour([0, clk_of_hour])
                bits_hour = self.hour_timer(clk_of_hour)
                bits_tens_hour, bits_ones_hour = [bits_hour[0]], bits_hour[1:] # The 1-bit high digit, The 4-bit low digit
                
                clk_of_am_pm = self.nand_gate_pm_am([bits_tens_hour[0], bits_ones_hour[-1]])
                current_am_pm_state = self.am_pm_memory.getQ_bar()
                am_pm_q, _ = self.am_pm_memory([current_am_pm_state, clk_of_am_pm, 0, 0])
            
    def __call__(self, clk = 0):
        # bits_tens_min, bits_ones_min, bits_tens_sec, bits_ones_sec = self.min_sec_timer(clk)
        clk = self.switch_sec([0, clk])
        bits_tens_sec, bits_ones_sec = self.second_timer(clk)
        clk_of_min = self.nand_gate_min_sec([bits_tens_sec[0], bits_tens_sec[-1]])
        clk_of_min = self.switch_min([0, clk_of_min])
        bits_tens_min, bits_ones_min = self.minute_timer(clk_of_min)
        
        clk_of_hour = self.nand_gate_hour_min([bits_tens_min[0], bits_tens_min[-1]])
        clk_of_hour = self.switch_hour([0, clk_of_hour])
        bits_hour = self.hour_timer(clk_of_hour)
        bits_tens_hour, bits_ones_hour = [bits_hour[0]], bits_hour[1:] # The 1-bit high digit, The 4-bit low digit
        
        clk_of_am_pm = self.nand_gate_pm_am([bits_tens_hour[0], bits_ones_hour[-1]])
        current_am_pm_state = self.am_pm_memory.getQ_bar()
        am_pm_q, _ = self.am_pm_memory([current_am_pm_state, clk_of_am_pm, 0, 0])
        
        return am_pm_q, bits_tens_hour, bits_ones_hour, bits_tens_min, bits_ones_min, bits_tens_sec, bits_ones_sec
```

```python
# Power on the V2 System Clock
system_clock = Oscillator()
manual_clock = Hour_Minute_Second_Timer_V2()

filename = "manual_time_setting_test.txt"

with open(filename, "w") as f:
    f.write("Action       | Time (HH:MM:SS AM/PM) | Raw Bits\n")
    f.write("-" * 65 + "\n")

    def log_state(action_text, am_pm, t_hr, o_hr, t_min, o_min, t_sec, o_sec):
        val_t_hr = bit_list_to_int(t_hr, signed=False)
        val_o_hr = bit_list_to_int(o_hr, signed=False)
        val_t_min = bit_list_to_int(t_min, signed=False)
        val_o_min = bit_list_to_int(o_min, signed=False)
        val_t_sec = bit_list_to_int(t_sec, signed=False)
        val_o_sec = bit_list_to_int(o_sec, signed=False)
        
        period = "AM" if am_pm == 0 else "PM"
        time_str = f"{val_t_hr}{val_o_hr}:{val_t_min}{val_o_min}:{val_t_sec}{val_o_sec} {period}"
        
        f.write(f"{action_text:12} | {time_str:20} | {am_pm} {t_hr}{o_hr} {t_min}{o_min} {t_sec}{o_sec}\n")

    # 1. Normal Ticking for 5 seconds (10 ticks)
    for _ in range(10):
        clk_signal = system_clock.level()
        state = manual_clock(clk_signal)
        if clk_signal == 1:
            log_state("Auto Tick", *state)
        system_clock.tick()
        
    f.write("\n--- USER RAPIDLY PRESSES 'HOUR' BUTTON 14 TIMES ---\n\n")

    # 2. Manual Override! Rapidly press the button 14 times
    manual_clock.update_hour(14)
    
    # Take a snapshot of the new time by passing clk=0 (no tick)
    state = manual_clock(0) 
    log_state("Btn Pressed", *state)

    f.write("\n--- USER RELEASES BUTTON, NORMAL CLOCK RESUMES ---\n\n")

    # 3. Normal Ticking for 5 more seconds to prove it isn't broken
    for _ in range(10):
        clk_signal = system_clock.level()
        state = manual_clock(clk_signal)
        if clk_signal == 1:
            log_state("Auto Tick", *state)
        system_clock.tick()

print(f"Test complete! Check '{filename}' to see your manual override in action.")
```
