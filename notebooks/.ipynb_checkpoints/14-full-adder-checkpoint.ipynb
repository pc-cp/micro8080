```python
# The Basic Gates
class AND:
    def __init__(self, nin=2):
        self.nin = nin

    def __call__(self, x):
        assert len(x) == self.nin, f"Expected {self.nin} inputs"
        result = x[0]
        # Chain bitwise AND over all inputs
        for val in x[1:]:
            result &= val
        return result

class OR:
    def __init__(self, nin=2):
        self.nin = nin

    def __call__(self, x):
        assert len(x) == self.nin, f"Expected {self.nin} inputs"
        result = x[0]
        # Chain bitwise OR over all inputs
        for val in x[1:]:
            result |= val
        return result

class Inverter:
    """The NOT gate, It strictly takes 1 input."""
    def __init__(self):
        self.nin = 1

    def __call__(self, x):
        assert len(x) == self.nin, "Inverter takes exactly 1 input"
        # If 1, return 0. If 0, return 1.
        return 1 - x[0]
```

```python
my_and = AND(nin=2)
print("AND Gate with [0, 0]:", my_and([0, 0]))
print("AND Gate with [0, 1]:", my_and([0, 1]))
print("AND Gate with [1, 0]:", my_and([1, 0]))
print("AND Gate with [1, 1]:", my_and([1, 1]))

my_or = OR(nin=2)
print("OR Gate with [0, 0]:", my_or([0, 0]))
print("OR Gate with [0, 1]:", my_or([0, 1]))
print("OR Gate with [1, 0]:", my_or([1, 0]))
print("OR Gate with [1, 1]:", my_or([1, 1]))

my_inverter = Inverter()
print("Inverter Gate with [0]:", my_inverter([0]))
print("Inverter Gate with [1]:", my_inverter([1]))
```

```python
# Composing Complex Gates
class NAND:
    def __init__(self, nin=2):
        self.nin = nin
        # NAND is just an AND gate followed by an Inverter
        self.and_gate = AND(nin)
        self.inverter = Inverter()

    def __call__(self, x):
        # x -> and -> invert -> out
        return self.inverter([self.and_gate(x)])

class NOR:
    def __init__(self, nin=2):
        self.nin = nin
        # NOR is just an OR gate followed by an Inverter
        self.or_gate = OR(nin)
        self.inverter = Inverter()

    def __call__(self, x):
        # x -> or -> invert -> out
        return self.inverter([self.or_gate(x)])
```

```python
# 1. Instantiate the "hardware"
my_nand = NAND(nin=2)
my_nor = NOR(nin=2)

print("NAND Gate with [0, 0]:", my_nand([0, 0]))
print("NAND Gate with [0, 1]:", my_nand([0, 1]))
print("NAND Gate with [1, 0]:", my_nand([1, 0]))
print("NAND Gate with [1, 1]:", my_nand([1, 1]))

print("NOR Gate with [0, 0]:", my_nor([0, 0]))
print("NOR Gate with [0, 1]:", my_nor([0, 1]))
print("NOR Gate with [1, 0]:", my_nor([1, 0]))
print("NOR Gate with [1, 1]:", my_nor([1, 1]))
```

```python
# Composing XOR gate use OR, NAND and AND gate
class XOR:
    def __init__(self):
        self.nin = 2

        self.or_gate = OR(self.nin)
        self.nand_gate = NAND(self.nin)
        self.and_gate = AND(self.nin)

    def __call__(self, x):
        assert len(x) == self.nin, "XOR takes exactly 2 inputs"
        # x -> or -> o1
        # x -> nand -> o2
        # and(o1, o2) -> out
        o1 = self.or_gate(x)
        o2 = self.nand_gate(x)
        return self.and_gate([o1, o2])
        
```

```python
my_xor = XOR()
print("XOR Gate with [0, 0]:", my_xor([0, 0]))
print("XOR Gate with [0, 1]:", my_xor([0, 1]))
print("XOR Gate with [1, 0]:", my_xor([1, 0]))
print("XOR Gate with [1, 1]:", my_xor([1, 1]))
```

```python
class HalfAdder:
    def __init__(self):
        self.nin = 2
        # The two components we need
        self.xor_gate = XOR()
        self.and_gate = AND(self.nin)

    def __call__(self, x):
        assert len(x) == self.nin, "Half Adder takes exactly 2 inputs"
        # Calculate Sum and Carry
        final_sum = self.xor_gate(x)
        final_carry = self.and_gate(x)

        # Return them as a tuple
        return final_carry, final_sum
```

```python
my_halfadder = HalfAdder()
print("HalfAdder with [0, 0]:", my_halfadder([0, 0]))
print("HalfAdder with [0, 1]:", my_halfadder([0, 1]))
print("HalfAdder with [1, 0]:", my_halfadder([1, 0]))
print("HalfAdder with [1, 1]:", my_halfadder([1, 1]))
```

```python
class FullAdder:
    def __init__(self):
        self.nin = 3 # # Carry-In, A, and B

        # A full adder is literally just two half adders and an OR gate
        self.ha1 = HalfAdder()
        self.ha2 = HalfAdder()
        self.or_gate = OR(2)

    def __call__(self, x):
        # Carry-In, A, and B
        if len(x) == self.nin:
            carry_in, a, b = x[0], x[1], x[2]
        else:
            carry_in = 0
            a, b = x[0], x[1]

        # First Half-Adder adds A and B
        carry1, sum1 = self.ha1([a, b])

        # Second Half-Adder adds previous sum and the Carry-In
        carry2, final_sum = self.ha2([carry_in, sum1])

        # The OR gate catches the carry from either Half-Adder
        final_carry = self.or_gate([carry2, carry1])

        # Return them as a tuple
        return final_carry, final_sum
```

```python
my_fulladder = FullAdder()
print("FullAdder with [0, 0, 0]:", my_fulladder([0, 0, 0]))
print("FullAdder with [0, 0, 1]:", my_fulladder([0, 0, 1]))
print("FullAdder with [0, 1, 0]:", my_fulladder([0, 1, 0]))
print("FullAdder with [0, 1, 1]:", my_fulladder([0, 1, 1]))
print("FullAdder with [1, 0, 0]:", my_fulladder([1, 0, 0]))
print("FullAdder with [1, 0, 1]:", my_fulladder([1, 0, 1]))
print("FullAdder with [1, 1, 0]:", my_fulladder([1, 1, 0]))
print("FullAdder with [1, 1, 1]:", my_fulladder([1, 1, 1]))
```

```python
class nBit_Adder:
    def __init__(self, nbits):
        self.nbits = nbits # Carry_In, A, and B
        # Instantiate 'n' Full Adders
        self.fulladders = [FullAdder() for _ in range(self.nbits)]

    def __call__(self, x, y, last_carry_in=0):
        assert len(x) == self.nbits and len(y) == self.nbits, f"Inputs must be {self.nbits} bits long"
        final_sums = []
        final_carrys = [last_carry_in]

        # Iterate backwards: start from the far right and move left
        for idx in reversed(range(self.nbits)):
            carry_in = final_carrys[0]
            a = x[idx]
            b = y[idx]
            
            # Run the Full Adder for this column
            carry_out, sum_out = self.fulladders[idx]([carry_in, a, b])
            
            final_carrys = [carry_out] + final_carrys
            final_sums = [sum_out] + final_sums
        # The final carry out of the MSB is still sitting at index 0
        return final_carrys[0], final_sums
```

```python
my_fulladder = nBit_Adder(1)
print("FullAdder with [0, 0, 0]:", my_fulladder([0], [0]))
print("FullAdder with [0, 0, 1]:", my_fulladder([0], [1]))
print("FullAdder with [0, 1, 0]:", my_fulladder([1], [0]))
print("FullAdder with [0, 1, 1]:", my_fulladder([1], [1]))
```

```python
def int_to_8bit_list(num):
    """Converts an integer to a 8-bit list with MSB at index 0."""
    # format(num, '08b') automatically converts to binary and pads to 8 bits
    bin_str = format(num, '08b')
    return [int(bit) for bit in bin_str]

def bit_list_to_int(bit_list):
    """Converts a bit list (MSB at index 0) back to a standard integer."""
    bin_str = "".join(str(bit) for bit in bit_list)
    return int(bin_str, 2)
```

```python
# Initialize 8-bit adder
adder8 = nBit_Adder(8)

def run_test_adder8(test_name, num1, num2):
    print(f"--- {test_name} ---")
    
    # Convert inputs to 16-bit hardware format
    x_bits = int_to_8bit_list(num1)
    y_bits = int_to_8bit_list(num2)

    # Run the hardware forward pass
    final_carry, sum_bits = adder8(x_bits, y_bits)
    # Convert result back to human-readable integer
    result_int = bit_list_to_int(sum_bits)

    # Display results
    print(f"  {num1} in binary: {x_bits}")
    print(f"+ {num2} in binary: {y_bits}")
    print("-" * 65)
    print(f"  Result bits:    {sum_bits}")
    print(f"  Final Carry Out: {final_carry}")
    print(f"  Decimal Result:  {result_int}")
    print(f"  Expected Sum:    {(num1 + num2) % 256}\n")
    print(f"  Expected Carry Out: {(num1 + num2) // 256}\n")

# ==========================================
# TEST CASES: STANDARD ADDITION
# ==========================================

# Test 1: Simple Addition (No major carry ripples)
# 15 (00001111) + 25 (00011001) = 40 (00101000)
run_test_adder8("Test 1: Simple Addition", 15, 25)

# Test 2: The Checkerboard (Testing every single pin)
# 170 (10101010) + 85 (01010101) = 255 (11111111)
run_test_adder8("Test 2: Checkerboard Bits", 170, 85)

# Test 3: The Full Ripple (Maximum 8-bit stress test)
# 255 (11111111) + 1 (00000001) = 0 (00000000) with a Carry Out of 1
run_test_adder8("Test 3: The 8-Bit Ripple Overflow", 255, 1)


# ==========================================
# ADVANCED HARDWARE: TWO'S COMPLEMENT SUBTRACTION
# ==========================================
# In hardware, Subtraction (A - B) is calculated as A + NOT(B) + 1.
# We can do this using your exact nBit_Adder by:
# 1. Inverting the bits of B (using the Inverter class we made earlier)
# 2. Setting 'last_carry_in' to 1.

my_inverter = Inverter()

def run_test_subtraction8(test_name, num_a, num_b):
    print(f"--- {test_name} ---")
    
    a_bits = int_to_8bit_list(num_a)
    b_bits = int_to_8bit_list(num_b)
    
    # 1. Pass B through an array of NOT gates (Inverting the bits)
    not_b_bits = [my_inverter([bit]) for bit in b_bits]
    
    # 2. Run the adder with the inverted B and a Carry-In of 1
    final_carry, diff_bits = adder8(a_bits, not_b_bits, last_carry_in=1)
    
    result_int = bit_list_to_int(diff_bits)
    
    print(f"  {num_a} in binary:       {a_bits}")
    print(f"- {num_b} in binary:       {b_bits}")
    print(f"  NOT({num_b}) in binary:  {not_b_bits}  <-- Inverted!")
    print("-" * 65)
    print(f"  Result bits:        {diff_bits}")
    print(f"  Decimal Result:     {result_int}")
    # Note: In subtraction, the final carry out is normally ignored 
    # (it indicates a positive result if 1, and negative if 0)
    expected_diff = num_a - num_b
    if expected_diff < 0:
         expected_diff = 256 + expected_diff # Wrap around for negative 8-bit numbers
    print(f"  Expected Diff:      {expected_diff}\n")


# Test 4: Basic Subtraction
# 50 - 18 = 32
run_test_subtraction8("Test 4: Subtraction (50 - 18)", 50, 18)

# Test 5: Subtracting a larger number from a smaller one
# 10 - 15 = -5 (Which in 8-bit Two's complement is 251 or 11111011)
run_test_subtraction8("Test 5: Negative Result (10 - 15)", 10, 15)
```

```python
def int_to_16bit_list(num):
    """Converts an integer to a 16-bit list with MSB at index 0."""
    # format(num, '016b') automatically converts to binary and pads to 16 bits
    bin_str = format(num, '016b')
    return [int(bit) for bit in bin_str]

# def bit_list_to_int(bit_list):
#     """Converts a bit list (MSB at index 0) back to a standard integer."""
#     bin_str = "".join(str(bit) for bit in bit_list)
#     return int(bin_str, 2)
```

```python
# 1. Instantiate the 16-bit hardware
adder16 = nBit_Adder(16)

def run_test(test_name, num1, num2):
    print(f"--- {test_name} ---")
    
    # Convert inputs to 16-bit hardware format
    x_bits = int_to_16bit_list(num1)
    y_bits = int_to_16bit_list(num2)

    # Run the hardware forward pass
    final_carry, sum_bits = adder16(x_bits, y_bits)
    # Convert result back to human-readable integer
    result_int = bit_list_to_int(sum_bits)

    # Display results
    print(f"  {num1} in binary: {x_bits}")
    print(f"+ {num2} in binary: {y_bits}")
    print("-" * 65)
    print(f"  Result bits:    {sum_bits}")
    print(f"  Final Carry Out: {final_carry}")
    print(f"  Decimal Result:  {result_int}")
    print(f"  Expected Sum:    {(num1 + num2) % 65536}\n")
    print(f"  Expected Carry Out: {(num1 + num2) // 65536}\n")

# ==========================================
# TEST CASES
# ==========================================

# Test 1: Basic Math
# Tests standard non-overlapping bit addition
run_test("Test 1: Standard Addition", 1024, 2048)

# Test 2: Alternating Bits (The Checkerboard Test)
# 43690 is 1010101010101010
# 21845 is 0101010101010101
# Adding them should result in 1111111111111111 (65535) with NO carry.
run_test("Test 2: Alternating Bits", 43690, 21845)

# Test 3: The Ultimate Ripple (Maximum Overflow)
# 65535 is 1111111111111111. Adding 1 requires the carry bit to successfully 
# ripple through all 16 Full Adders from right to left. 
# Result should be 0000000000000000 with a Final Carry Out of 1.
run_test("Test 3: The 16-Bit Ripple Overflow", 65535, 1)
```

```python
# 1. Instantiate the 16-bit hardware
adder8_low_byte = nBit_Adder(8)
adder8_high_byte = nBit_Adder(8)

def run_test(test_name, num1, num2):
    print(f"--- {test_name} ---")
    
    # Convert inputs to 16-bit hardware format
    x_bits = int_to_16bit_list(num1)
    y_bits = int_to_16bit_list(num2)

    # Run the hardware forward pass
    sum_bits = []
    
    carry_low, sum_bits_low = adder8_low_byte(x_bits[8:], y_bits[8:])
    final_carry, sum_bits_high = adder8_high_byte(x_bits[0:8], y_bits[0:8], carry_low)
    sum_bits = sum_bits_high + sum_bits_low
    # Convert result back to human-readable integer
    result_int = bit_list_to_int(sum_bits)

    # Display results
    print(f"  {num1} in binary: {x_bits}")
    print(f"+ {num2} in binary: {y_bits}")
    print("-" * 65)
    print(f"  Result bits:    {sum_bits}")
    print(f"  Final Carry Out: {final_carry}")
    print(f"  Decimal Result:  {result_int}")
    print(f"  Expected Sum:    {(num1 + num2) % 65536}\n")
    print(f"  Expected Carry Out: {(num1 + num2) // 65536}\n")

# ==========================================
# TEST CASES
# ==========================================

# Test 1: Basic Math
# Tests standard non-overlapping bit addition
run_test("Test 1: Standard Addition", 1024, 2048)

# Test 2: Alternating Bits (The Checkerboard Test)
# 43690 is 1010101010101010
# 21845 is 0101010101010101
# Adding them should result in 1111111111111111 (65535) with NO carry.
run_test("Test 2: Alternating Bits", 43690, 21845)

# Test 3: The Ultimate Ripple (Maximum Overflow)
# 65535 is 1111111111111111. Adding 1 requires the carry bit to successfully 
# ripple through all 16 Full Adders from right to left. 
# Result should be 0000000000000000 with a Final Carry Out of 1.
run_test("Test 3: The 16-Bit Ripple Overflow", 65535, 1)
```
